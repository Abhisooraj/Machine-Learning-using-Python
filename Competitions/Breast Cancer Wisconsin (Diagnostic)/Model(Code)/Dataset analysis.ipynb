```python
#Converting object column (diagnosis) into numerical column because ML algorithms does not take values other than numerical
# Assigining 1 to M and 0 to B.
def trans_diag(x):
    if x == 'M':
        return 1
    if x == 'B':
        return 0
```

```python
cancer_data['diagnosis'] = cancer_data['diagnosis'].apply(trans_diag)
```

```python
cancer_data.info()
```

```python
#We compare the correlation between features and remove one of two features that have a correlation higher than 0.9
cancer_corr = cancer_data.corr()['radius_mean']
cancer_corr
```
